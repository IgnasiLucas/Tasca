```r
#En primer lugar, preparamos el ambiente:
system2(command = './preparar_ambiente.sh', wait = TRUE)
```

```r
#Hemos creado el objeto nuevo con la proteína CHMR1.fas, con un e muy restrictivo 
BlastpOutCHRM1 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRM1.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)
```

```r
#En la tabla hemos metido todo lo que coincide
TablaOutCHRM1 <- read.table(textConnection(BlastpOutCHRM1),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))
```

```r
#Dimensiones de la tabla
dim(TablaOutCHRM1)
```

```r
#Visualizamos la tabla
TablaOutCHRM1
```

```r
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

# La función lapply() aplicará una función sobre los e listados. A mayor e, menos restrictividad y por tanto, más resultados esperamos
Lista_de_Tablas_CHRM1 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut2 <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRM1.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut2),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
#Mostramos las seis primeras filas de la sexta tabla creada en el paso anterior
head(Lista_de_Tablas_CHRM1[[6]])
```

```r
#Observamos las secuencias homólogas encontradas con la función dim() aplicada 
#a la primera dimensión (número de filas)
#Vemos disminuir la restricción
Numero_de_resultados_CHRM1 <- sapply(Lista_de_Tablas_CHRM1, function(x) dim(x)[1])
Numero_de_resultados_CHRM1
```

```r
#Representamos gráficamente:
plot(Valores_E_maximos, Numero_de_resultados_CHRM1, log = 'xy', type = 'b')
```

```r
# La función segments() añade segmentos a un gráfico previo. Por eso, represento
# antes un gráfico con los rangos adecuados, pero vacío (type='n'). Para determinar
# el rango del eje horizontal, determino primero el valor máximo de "qend".

#Se muestran los resultados de las tablas 1 y 5.
```

```r
inicios <- Lista_de_Tablas_CHRM1[[5]]$qstart
finales <- Lista_de_Tablas_CHRM1[[5]]$qend
valoresE <- Lista_de_Tablas_CHRM1[[5]]$evalue


#A menor e, menor número de resultados

AlineamientoMaximoCHRM1 <- max(Lista_de_Tablas_CHRM1[[5]]$qend)
plot(c(0, AlineamientoMaximoCHRM1), range(valoresE), type='n', xlab='Posición', ylab='Valor E')
segments(inicios, valoresE, finales, valoresE)
```

```r
#Vemos como la tabla 1 presenta menos valores que la 5. Vemos más homologías en las primeras posiciones.

inicios <- Lista_de_Tablas_CHRM1[[1]]$qstart
finales <- Lista_de_Tablas_CHRM1[[1]]$qend
valoresE <- Lista_de_Tablas_CHRM1[[1]]$evalue


#A menor e, menor número de resultados

AlineamientoMaximoCHRM1 <- max(Lista_de_Tablas_CHRM1[[1]]$qend)
plot(c(0, AlineamientoMaximoCHRM1), range(valoresE), type='n', xlab='Posición', ylab='Valor E')
segments(inicios, valoresE, finales, valoresE)
```

```r
#Finalmente exploramos la distribución taxonómica.
#La columna sblastname ofrece una clasificación fácilmente interpretable de las secuencias, las cuales contamos.
#Se muestran las encontradas en cada grupo taxonómico aplicando la función table() a esa columna de cada tabla en Lista_de_Tablas:ómica.
```

```r
lapply(Lista_de_Tablas_CHRM1, function(x) table(x$sblastname))
```

```r
#Para ver la relación entre la longitud del alineamiento y la distribución taxonómica, 
#repetimos el recuento pero contando solo al menos 300 aa.

lapply(Lista_de_Tablas_CHRM1, function(x) {
       filtro <- x$length >= 300
       table(x[filtro, 'sblastname'])
    }
)
```
