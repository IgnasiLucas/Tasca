```r
#En primer lugar, preparamos el ambiente:
system2(command = './preparar_ambiente.sh', wait = TRUE)
```

```r
#Hemos creado el objeto nuevo con la proteína CHMR1.fas, con un e muy restrictivo 
BlastpOutCHRNA7 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA7.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)
```

```r
#En la tabla hemos metido todo lo que coincide
TablaOutCHRNA7 <- read.table(textConnection(BlastpOutCHRNA7),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))
```

```r
#Dimensiones de la tabla
dim(TablaOutCHRNA7)
```

```r
#Visualizamos la tabla
TablaOutCHRNA7
```

```r
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

# La función lapply() aplicará una función sobre los e listados. A mayor e, menos restrictividad y por tanto, más resultados esperamos
Lista_de_Tablas_CHRNA7 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut4 <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRNA7.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut4),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
#Mostramos las seis primeras filas de la sexta tabla creada en el paso anterior
head(Lista_de_Tablas_CHRNA7[[6]])
```

```r
#Observamos las secuencias homólogas encontradas con la función dim() aplicada 
#a la primera dimensión (número de filas)
#Vemos disminuir la restricción
Numero_de_resultados_CHRNA7 <- sapply(Lista_de_Tablas_CHRNA7, function(x) dim(x)[1])
Numero_de_resultados_CHRNA7
```

```r
#Representamos gráficamente:
plot(Valores_E_maximos, Numero_de_resultados_CHRNA7, log = 'xy', type = 'b')
```
