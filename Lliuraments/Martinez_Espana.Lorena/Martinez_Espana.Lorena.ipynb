# 3.BLAST (Psi-blast)

En este apartado sólo vamos a trabajar con la secuencia de una proteína, que será CHRNA7, aunque realmente podríamos haber utilizado también
la secuencia de CHRNA3 o CHRM1. Lo que vamos a hacer es poner en práctica las herramientas aprendidas de blast y, en particular, psi-blast. 

Nuestro objetivo principal es ver la distribución taxonómica de esta proteína y observar en qué organismos se encuentra y cuán diferentes 
son entre ellas las proteínas homólogas. Para buscar estas proteínas utilizamos la base de datos de Swissprot, la cual hemos utilizado durante 
las prácticas de la asignatura y la cual tenemos ya cargada en el espacio de trabajo. Tenemos cargados algunos organismos en los que podría encontrarse homólogos a nuestra proteína y es lo que vamos a estudiar. 

Durante el desarrollo de este apartado se irán ejecutando diferentes comandos que iremos explicando

```r
#Primero preparamos el espacio de trabajo, instalando blast y descomprimiendo los archivos necesarios
system2(command = './preparar_ambiente.sh', wait = TRUE)
```

Tras preparar el espacio de trabajo, vamos a buscar si existe alguna proteína con cierta homología a CHRNA7 entre las proteínas
que nos proporciona la base de datos con los organismos de los que disponemos. Vamos a comenzar siendo muy poco restrictivos, es decir, 
en 'evalue' vamos a poner un valor de 1.0e-10, lo cual hará que nos aparezcan bastantes proteínas ya que no estamos pidiendo
que tengan un nivel de homología muy alto. Posteriormente, guardaremos esos datos en un archivo html (`blastp1.html`) y podremos observar esas proteínas, su secuencia y alguna información extra.

Antes de realizar esta tarea, hemos de aclarar que el 'evalue' o el 'valor E' es un valor numérico elegido por nosotros mismos y que restringe las búsqueda de proteínas los diferentes organismos disponibles. Las restricciones hacen referencia al grado de homología, al grado de parecido que queremos que tengan las proteínas que se buscan en la base de datos, por lo que cuanto menor sea el valor, cuanto más se acerque a cero, más "significante" será ese match. Realmente, el valor E mide cuántas proteínas podemos encontrar por casualidad cuando buscamos en una base de datos de cierto tamaño, midiendo el ruído de fondo, pero es una cuestión con cálculos complejos en la que no vamos a meternos.

```r
blastp1 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                               '-query', 'CHRNA7.fas',
                               '-evalue', '1.0e-10',
                               '-html',
                               '-out', 'blastp1.html'),
                      wait = TRUE,
                      stdout = TRUE)
```

Si vamos a ese archivo html podemos ver proteínas homólogas a CHRNA7 presentes en diferentes organismos como
*Mus musculus* o *Gallus gallus*. En el archivo también nos dicen el tipo de ajuste que se ha realizado, que es 'Compositional matrix adjust',
además de otra información accesoria. 

Pero si queremos ver los datos con más claridad para poder realizar diferentes análisis, podemos representar los datos en forma de tabla
y saber las dimensiones que ésta tendría. Para ello, primero, crearíamos un nuevo archivo, pero en este caso de texto plano, para poder organizar
la información en forma de tabla

Nosotros hemos cambiado el nombre a las columnas para poder trabajar la información de manera más sencilla y poder identificar la información que nos porporciona cada una de las columnas de la tabla.

```r
#Creamos un archivo de texto plano

blastptexto <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA7.fas',
                              '-evalue', '1.0e-10',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

#Transformamos el texto plano en un data frame y obtenemos una tabla con toda la información distribuida en columnas

Tabla1 <- read.table(textConnection(blastptexto),
                        sep = '\t',
                        col.names = c('Identificador', 'porcentajeIdentidad', 'Longitud', 'PrimeraPosiciónQ',
                                     'ÚltimaPosiciónQ', 'PrimeraPosiciónS', 'ÚltimaPosiciónS', 'ValorE', 'IdentificadorS',
                                     'NombreCientífico', 'Taxón'))
Tabla1

#Dimensiones de la tabla
dim(Tabla1)
```

En la tabla podemos observar la información de forma mucho más clara y en cada columna tenemos una información, a la cual hemos nombrado
en castellano y con un nombre más claro cambiando los 'col.names' en los códigos. Cuando hemos creado el archivo de texto se observan los nombres
originales que tenían las columnas, pero los hemos cambiado para poder trabajar más cómodamente, el significado de cada head de columna es el siguiente:

- **Identificador:** número de acceso o identificador de la secuencia encontrada (subject).
- **porcentajeIdentidad:** porcentaje de identidad entre la query y el subject en su alineamiento local.
- **Longitud:** longitud del alineamiento entre query y subject.
- **PrimeraPosiciónQ:** primera posición de la query alineada.
- **ÚltimaPosiciónQ:** última posición alineada en la query.
- **PrimeraPosiciónS:** primera posición de la subject alineada.
- **ÚltimaPosiciónS:** última posición de la subject alineada.
- **ValorE:** valor E del alineamiento. Es decir, número de alineamientos de igual o mejor puntuación esperados por azar en una base de datos igual de grande.
- **IdentificadorS:** identificador de la especie a la que pertenece la subject, en la base de datos de taxonomía del NCBI.
- **NombreCientífico:** nombre científico de la especie a la que pertenece la subject.
- **Taxón:** nombre de un grupo taxonómico de rango superior al que pertenece la subject, usado para facilitar la interpretación.

También vemos las dimensiones de la tabla, la cual nos presenta 160 proteínas homólogas siguiendo el valor E que nosotros le hemos asignado, 
que era muy poco restrictivo. 

A continuación vamos a realizar el mismo proceso pero con un valor E mucho menor para restringuir la búsqueda a proteínas con un alto grado
de homología.

```r
#Creamos un archivo de texto plano

blastptexto2 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA7.fas',
                              '-evalue', '1.0e-150',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

#Transformamos el texto plano en un data frame y obtenemos una tabla con toda la información distribuida en columnas

Tabla2 <- read.table(textConnection(blastptexto2),
                        sep = '\t',
                        col.names = c('Identificador', 'porcentajeIdentidad', 'Longitud', 'PrimeraPosiciónQ',
                                     'ÚltimaPosiciónQ', 'PrimeraPosiciónS', 'ÚltimaPosiciónS', 'ValorE', 'IdentificadorS',
                                     'NombreCientífico', 'Taxón'))
Tabla2

#Dimensiones de la tabla
dim(Tabla2)
```

Ahora hemos querido restringuir mucho nuestra búsqueda y hemos aplicado un valor E de 1.0e-150, lo cual hace que nuestra búsqueda de proteínas
en diferentes taxones se reduzca a 7 proteínas, las cuales pueden observarse en la nueva tabla creada (Tabla2). 

Hemos de añadir que el valor E aplicado dependerá del tipo de estudio que queramos realizar y del grado de homología que necesitemos. 

Sin embargo, si queremos ver las diferentes proteínas que se encuentran en Swissprot para ciertos valroes E, no necesitamos ejecutar los comandos que hemos comentado con anterioridad, sino que podemos establecer una serie de valores máximos y crear tablas diferentes para cada uno de ellos, y mostrarlas según necesitemos. 

Es de suponer que, cuanto más pequeño sea el valor que apliquemos, menos proteínas mostraran homología y al revés, cuanto mayor sea
el valor E, más proteínas homólogas encontraremos en la base de datos. 

```r
#Establecemos un conjunto de valores máximos (valores E)

Valores_E_escogidos <- c('1.0e-150', '1.0e-100', '1.0e-70', '1.0e-50', '1.0e-20',
                       '1.0e-15', '1.0e-10', '1.0e-07', '1.0e-05', '1.0e-03', '1.0e-02', '1', '5')


#A continuación, creamos un conjunto de tablas con los valores E contenidos en Valores_E_escogidos
Todas_las_tablas <- lapply(Valores_E_escogidos,
                          function(x) {
                              BlastpOut <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRNA7.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut),
                                         sep = '\t',
                                         col.names = c('Identificador', 'porcentajeIdentidad', 'Longitud', 'PrimeraPosiciónQ',
                                     'ÚltimaPosiciónQ', 'PrimeraPosiciónS', 'ÚltimaPosiciónS', 'ValorE', 'IdentificadorS',
                                     'NombreCientífico', 'Taxón'))})
```

Lo que hemos hecho ha sido como juntar diferentes ejecuciones de blastp en unos pocos comandos y almacenar la información de cada uno
en una tabla que tendrán un número. El número corresponderá al orden que tengan su valor E en la función Valores_E_escogidos. 

Podemos observar la tabla que nosotros queramos (la que tenga el valor E que deseamos) ejecutando los siguientes comandos

```r
#Mostramos la tabla que nos interesa

tail(Todas_las_tablas [[13]])
```

Cambiando el número que está entre los corchetes, que corresponde al blastp que tiene el valor E número 13, podemos sacar los datos
que correspondan a cualquier otro blastp. 

También podemos ver cuántas proteínas encontramos según el valor E ejecutado, sin necesidad de hacerlo valor por valor, ejecutando
el siguiente comando

```r
Numero_de_proteínas <- sapply(Todas_las_tablas, function(x) dim(x)[1])
Numero_de_proteínas
```

Tal como hemos mencionado en los párrafos anteriores, cuanto mayor es el valor E, más proteínas se encuentran en las bases de datos
que presenten algún tipo de homología, pudiendo representar esos datos de forma gráfica para ver cuántas proteínas se podrían encontrar
según los valores E. 

```r
#Realizamos una representación donde en el eje x se muestran los valores E y en el eje y el número
#de proteínas homólogas encontradas, representadas en forma logarítmica

plot(Valores_E_escogidos, Numero_de_proteínas, log = 'xy', type = 'b')
```

Al final todas estas herramientas nos ofrecen diferentes opciones para ver cuantas proteínas se encuentran según el valor E que apliquemos, lo cual nos da una información general sobre la distribución de esta proteína en los organismos  según el grado de parecido que queramos. 

Sin embargo, hemos de tener en cuenta que CHRNA7 es una proteína de 502 aminoácidos y que algunas de las proteínas que nos aparecen en los resultados anteriores pueden tener homología sólo con ciertas regiones de nuestra proteína. Para ello, vamos a realizar una representación para ver, según el valor E, la longitud del alineamiento que encontramos. 

```r
#Mediante estos códigos conseguiremos realizar una representación
#donde en el eje y se representan los valores E y en el eje x
#la longitud de los alineamientos realizados por BLAST.
#Los datos de longitud se obtienen mediante el símbolo '$'
plot(Todas_las_tablas[[4]]$Longitud, Todas_las_tablas[[4]]$ValorE,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

Hemos utilizado el valor E número 4, es decir, un valor E de 1.0e-50, lo cual es bastante restrictivo y podemos ver que la mayoría de proteína con cierto grado de homología que se encuentran en la base de datos tienen una longitud de alineamiento, es decir, se han alineado con una longitud de alrededor de 500 aminoácidos, que es casi la longitud de CHRNA7, por lo que estas proteínas serían interesante estudiarlas para ver si realmente tienen una relación real con nuestra proteína de estudio.

Sin embargo, vemos que existen ciertas proteínas con una longitud de alineamiento de 300-350 aminoácidos, por lo que presentarían una homología parcial con CHRNA7 y también son interesantes de estudiar ya que puedes ser proteínas que estén muy alejadas y tengan estructuras muy diferentes pero compartir cierto grado de su función. 

En esas proteínas nos vamos a centrar para ver cuál es la región que comparten con CHRNA7 y así poder investigar un poco si las proteínas pueden tener funciones similares en función de las regiones que comparten. 

```r
Comienzo <- Todas_las_tablas[[4]]$PrimeraPosiciónQ
Final <- Todas_las_tablas[[4]]$ÚltimaPosiciónQ
Valor_E <- Todas_las_tablas[[4]]$ValorE

#De nuevo, vamos a utilizar las proteínas con el valor E 4 (1.0e-50)
#Y utilizaremos la primera posición y la última posición del alineamiento
#y los valores E correspondientes para ver las regiones compartidas por las proteínas

AlineamientoNuevo <- max(Todas_las_tablas[[4]]$ÚltimaPosiciónQ)
plot(c(0, AlineamientoNuevo), range(Valor_E), type='n', xlab='Posición', ylab='ValorE')
segments(Comienzo, Valor_E, Final, Valor_E)
```

A pesar de tener muy pocas proteínas con homología, podemos ver que lo que se comparte, principalmente, es la región central la que comparten todas estas proteínas.

## Conclusiones y discusión

Tras realizar el análisis con BLAST, podemos sacar numerosas conclusiones. La primera es que suponemos que nuestros resultados serían válidos con las conclusiones aportadas por *Viscardi et al.* en 2021, en las que señalaban que las proteínas relacionadas con la neurorrecepción nacieron en el momento de divergencia entre cordados y  cniadrios, excluyendo a organismos como hongos, plantas o procariotas, aunque en nuestros resultados podemos ver que se ha encontrado una proteína en enterobacterias, la cual tendría una función completamente diferente pero algun tipo de parecido en su estructura. Aunque, cabe destacar, que tampoco aparece en el grupo de los cnidarios, a pesar de que son organismos con sistema nervioso primitivo, por lo que cabría realizar estudios de filogenia con más proteínas que conformen los receptores nicotínicos para ver por qué estas proteínas no aparecen en cnidarios, a pesar de ser el momento de divergencia entre cordados y cnidaria cuando parecen tener su origen evolutivo. 

Sin embargo, sí que se encuentran proteínas homólogas en diferentes grupos de cordados como los pájaros, los peces, roedores y primates, organismos con un sistema nervioso bastante desarrollado y, en cierta manera, parecido al nuestro, aunque menos sofisticado y carente de algunas funciones cognitivas. Otro aspecto a comentar es que, aun aplicando condiciones de búsquedas muy restrictivas, aparecen proteínas homólogas en nematodos, animales pseudocelomados con un sistema nervioso mucho más simple y atrasado que el de los animales superiores, como son los mamíferos. Sin embargo, se necesitarían otros tipo de estudios para ver qué tipo de función tienen esas proteínas homólogas en estos animales, realizados en laboratorios relacionados con la zoología y la fisiología animal.

Por último, quería señalar que en roedores y primates se encuentra gran número de proteínas homólogas a CHRNA7, resultado que parece bastante acertado debido a que los roedores y los primates son animales muy utilizados en los estudios del sistema nerviosos debido al parecido que muestra su sistema y el nuestro, además que han mostrado algunos de rasgos de capacidad cognitiva.
