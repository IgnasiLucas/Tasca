## Introducción
El objetivo de este análisis es explorar la distribución taxonómica de las secuencias homólogas a la de la proteína humana CHRM1 (receptor muscarínico de acetilcolina). Debido a que apareció entre los linajes de cnidarios y cordados, en este análisis no deberíamos encontrar homologías en ctenóforos, poríferos, placozoa, hongos o plantas.

## Métodos
La base de datos de secuencias empleada es Swissprot, porque es una base de datos de proteínas ya instalada localmente en este ambiente.

Se utiliza **blastp** (Altschul *et al.*, 1990) con diferentes umbrales de valor E, para determinar a qué ritmo aumenta la distribución taxonómica de las secuencias encontradas a medida que se relaja el grado de similitud exigido. 

## Resultados
### BLASTP individual
Se ejecuta blastp, con umbral de valor E de 1.0e-60, que es muy exigente, para obtener secuencias parecidas a la original. Se presentan los resultados en forma de tabla.

```r
BlastCHRM1 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRM1.fas',
                              '-evalue', '1.0e-60',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

TablaCHRM1blast <- read.table(textConnection(BlastCHRM1),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))

dim(TablaCHRM1blast)
```

```r
TablaCHRM1blast
```

La tabla de resultados incluye 34 secuencias encontradas, incluyendo la misma secuencia usada como *query*. Las columnas de la tabla són:

- **saccver**: número de acceso o identificador de la secuencia encontrada (*subject*).
- **pident**: porcentaje de identidad entre la *query* y el *subject* en su alineamiento local.
- **length**: longitud del alineamiento entre *query* y *subject*.
- **qstart**: primera posición de la *query* alineada.
- **qend**: última posición alineada en la *query*.
- **sstart**: primera posición de la *subject* alineada.
- **send**: última posición de la *subject* alineada.
- **evalue**: valor E del alineamiento. Es decir, número de alineamientos de igual o mejor puntuación esperados por azar en una base de datos igual de grande.
- **staxid**: identificador de la especie a la que pertenece la *subject*, en la base de datos de taxonomía del NCBI.
- **ssciname**: nombre científico de la especie a la que pertenece la *subject*.
- **sblastname**: nombre de un grupo taxonómico de rango superior al que pertenece la *subject*, usado para facilitar la interpretación.

### BLASTP en serie

```r
Valores_E_maximos <- c('1.0e-60', '1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

Tablas_CHRM1 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRM1.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
tail(Tablas_CHRM1[[6]])
```

A continuación, se presentan el número de filas obtenidas en el blastp para cada valor de e: 

```r
Numero_de_resultados <- sapply(Tablas_CHRM1, function(x) dim(x)[1])
Numero_de_resultados
```

```r
#Se representan en una gráfica los valores de e en función al número de resultados obtenidos en blastp. 
plot(Valores_E_maximos, Numero_de_resultados, log = 'xy', type = 'b')
```

A continuación, se examina la relación entre la longitud del alineamiento y el valor E de los resultados guardados en la última tabla, la número 11, en la que se ha usado un valor E de 1 y que contiene un mayor número de resultados. Como cabe esperar, a mayor valor de E, mayor número de resultados. 

```r
plot(Tablas_CHRM1[[11]]$length, Tablas_CHRM1[[11]]$evalue,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

Aunque la proteína CHRM1 es de unos 460 aminoácidos, existen alineamientos de menos de 100. Esto indica que existen muchas proteínas en la base de datos que presentan homología **parcial** con CHRM1. Es decir, que muchos de estos resultados corresponden a proteínas que solo se parecen a CHRM1 en un fragmento concreto; probablemente un dominio de función conservada. Cabe preguntar qué posiciones de CHRM1 son esas que participan en los alineamientos parciales. 

```r
inicios <- Tablas_CHRM1[[11]]$qstart
finales <- Tablas_CHRM1[[11]]$qend
valoresE <- Tablas_CHRM1[[11]]$evalue

AlineamientoMaximo <- max(Tablas_CHRM1[[11]]$qend)
plot(c(0, AlineamientoMaximo), range(valoresE), type='n', xlab='Posición', ylab='Valor E')
segments(inicios, valoresE, finales, valoresE)
```

Esta gráfica muestra que la homología encontrada con valores E mayores no está distribuída aleatoriamente a lo largo de la secuencia de CHRM1, sino que, efectivamente, se concentra entre las posiciones 300 y 400, aproximadamente.

Por último, se va a explorar la distribución taxonómica de las secuencias encontradas. 

```r
lapply(Tablas_CHRM1, function(x) table(x$sblastname))
```

El resultado del código anterior es una lista de recuentos de los diferentes grupos taxonómicos presentes en cada tabla de resultados de **blastp**. Se observa, por ejemplo, que en los resultados del sexto **blastp** (umbral de valor E de $10^{-10}$) empiezan a aprecer secuencias homólogas a CHRM1 procedentes de virus.

Viscardi et al. (2021) alegan que los receptores de acetilcolina aparecieron en el linaje del último ancestro común entre cordados y cnidarios, por lo que cabría esperar que pudiesen haber homologías dentro de los diferentes filos a partir de cnidarios. Las proteínas homólogas con valores E menores y con alineamientos más completos pertenecerán a animales más próximos filogenéticamente a los humanos. Para ver la relación entre la longitud del alineamiento y la distribución taxonómica,  se aplicará un filtro al recuento de grupos taxonómicos pero contando solamente las secuencias con un alineamiento de al menos 300 aminoácidos:

```r
lapply(Tablas_CHRM1, function(x) {
       filtro <- x$length >= 300
       table(x[filtro, 'sblastname'])
    }
)
```

## Discusión
Utilizando la secuencia de la proteína CHRM1 de humanos como consulta (*query*) en búsquedas de **blastp**, encontramos proteínas homólogas en multitud de especies. También se confirma que no existen homólogos en ctenóforos, poríferos, placozoa, hongos o plantas. Existe una región en la distribución de los alineamientos a lo largo de la secuencia CHRM1, entre los 300 y 400 aminoácidos, que es donde se concentra la homología entre las proteínas. Se han obtenido resultados no esperables (como homologías en virus). Esto puede deberse a que, aumentando el valor de E, las búsquedas son menos específicas. Como se puede ver, al aplicar el filtro de compartir al menos 300 aa, los virus ya no aparecen en la lista.

Una búsqueda manual de la secuencia en la base de datos Pfam (Mistry *et al*, 2021) muestra que se trata de la *familia de receptores acoplados a proteína G similares a la Rodopsina* ( IPR000276). Incluye hormonas, neurotransmisores y receptores de luz y todos son transductores de señales extracelulares a través de la interacción con proteínas de unión a nucleótidos de guanina (G).Aunque varían en tipo de ligando, las secuencias de aminoácidos de los receptores son muy parecidos, adoptando una estructura de 7 hélices transmembrana. El receptor CHRM1 pertenece a la subfamilia A18.
