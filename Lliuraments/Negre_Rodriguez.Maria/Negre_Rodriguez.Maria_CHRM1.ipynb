```r
#Nuestro objetivo es analizar cada proteína propuesta y poder con ayuda del programa, entender de dónde viene esa proteína, 
# ancestro común y analizar su expresión. Tanto Blast como los paquetes ya están instalados.
#Basándonos en el modelo de ejemplo propuesto, nosotros queremos primero, estudiar lo que contiene nuestro paquete CHRM1.fas, al abrirlo, 
# vemos una secuencia de aminoácidos. Nosotros, queremos ver que seres vivos presentan esta secuencia de aminoácidos. 
# Primero, realizaré un análisis restrictivo de mi base de datos para así estudiar los alineamientos. Lo voy a guardar en BlastpOutCHRM1. 
# Además, marcaré el nombre y contenido de cada columna. 
BlastpOutCHRM1 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRM1.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

# Voy a crear un tabla, llamada CHRM1, de mi proteína CHRM1. Donde me haga un análisis con las especies que 
# contengan esta secuencia de aminoácidos. Ahora, debo estudiar el alineamiento de las secuencias en condiciones más o menos restrictivas,
# en este primer caso, estableceré un valor de E muy pequeño, como se ve en la parte superior: E = 1.0e-50.
# A menor E, mayor restricción.
# Ahora creo una tabla con los alinemientos y estudio los diferentes parámetros:
TablaOutCHRM1 <- read.table(textConnection(BlastpOutCHRM1),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))

dim(TablaOutCHRM1) #Con esta orden, le estoy pidiendo que me diga las dimensiones de la tabla.
```

```r
#Una vez creado el objeto BlastpOUTCHRM1 y TablaOUTCHRM1, me interesa estudiar esa tabla obtenida. 
```

```r
# Ahora veremos la tabla que se me ha formado, llamada CHRM1
TablaOutCHRM1
# en la tabla podemos ver como la secuencia que teníamos de aminoácidos de CHRM1, en condiciones restrictivas, se ha alineado y coincide con las secuencias
# de las siguientes especies: primates, roedores, pájaros, nemátodos, curstáceos... Observamos que esta proteína sobretodo coincide con secuencias en 
# primates y roedores, siendo éstos genomas muy parecidos al humano. El ejercicio nos indicada que CHRM1, era una proteína humana, sabiendo que el genoma
# de roedor coincide en un 95% con el humano, y los primates un 99% con el humano, no nos debemos extrañar con estos daos. 
# Gracias a esta tabla podemos ver como la CHRM1, el receptor muscarínico de la acetilcolina, es una proteína que contienen varias especies que estan relacionadas
#evolutivamente, es decir, que hubo un ancestro común que poseía una proteína parecida a ésta y de la cual han divergido las diferentes especies y por ello,
#esta proteína humana, muestra semejanzas con otras que ejercen una función parecida en otras especies relacionadas. Pero, no podemos saber hasta que punto
# esta proteína se parece más o menos a la humana en comparación con otras especies solamente con esta tabla, debemos realizar más análisis.
```

```r
# ¿Cómo analizamos mejor estos alineamientos? Una forma de hacerlo, es realizar una tabla "seriada" es decir, nosotros, vamos a crear tablas como la de arriba,
# En este caso, crearemos 10 tablas, pero entre ellas diferirán. Van a diferir en lo restrictivas que sean, es decir, en su valor e. Como podemos observar,
# Vamos a partir de una tabla (1) con un e muy restrictivo y acabaremos con la tabla (10) con un e muy laxo. Con esto, de cierta manera estoy creando 
# una especie de "árbol" filogenético con datos. Aquellas especies que aparezcan en la primera tabla, estarán más relacionadas con la proteína humnaa 
#evolutivamente, a medida que aumente e irán apareciendo más y más especies que se siguen alineando con la secuencia de mi proteína, pero de una forma
# más laxa. Lo vamos a ver paso por paso a continuación:
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

# lapply(), me permite establecer una función sobre cada valor e que he establecido anteriormente. En esta función, nosotros estamos diciéndole al programa
# que queremos que me analice BlastpOUT_CHRM1, con la base swissprot me encuentre las especies que tienen secuencias de aminoácidos que se alineen con la
# secuencia que tengo en mi archivo CHRM1.fas. Y además quiero que me genere una tabla que analice diferentes parámetros que le he indicado que 
# sean columnas: longitud, pendiente, la especie y la familia... 
Lista_de_Tablas_CHRM1 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut_CHRM1 <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRM1.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut_CHRM1),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
# Este objeto que he creado, llamado lista_de_tablas_CHRM1, me va a guardar las diez tablas que he generado, resultado de los diez blastp que he ejecutado.
#Si quiero ver una de las tablas generadas, puedo ver una tabla individual y además, ver únicamente de esa tabla, los últimos resultados con la función
# tail.
tail(Lista_de_Tablas_CHRM1[[5]])
```

```r
# Ahora, además, puedo de cada tabla generada con diferentes valores de e generada para blastp en la base de datos swissprot; puedo decirle al programa
#que me enseñe cada tabla, sus dimensiones, es decir que me diga el número de filas y columnas que se ha generado; aplico la función dim(), y me quedo con 
# la primera dimensión "X". Con estos resultados, podemos observar como, a medida que aumentamos el p-valor, 
# aumentan los valores obtenidos. Con un E con valor 1, me da 774 filas, esto es porque conforme aumente el valor E, más filas obtendré, más laxos son los 
#alineamientos.
Numero_de_resultados_CHRM1 <- sapply(Lista_de_Tablas_CHRM1, function(x) dim(x)[1])
Numero_de_resultados_CHRM1
```

```r
#Puedo además, analizar de una forma más "visual" los resultados obtenidos mediante una gráfica. 
plot(Valores_E_maximos, Numero_de_resultados_CHRM1, log = 'xy', type = 'b')
```

```r
# Ahora, puedo además, estudiar las diferencias en las longitudees de alineamiento a medida que aumenta el valor de e. Puedo ir tabla a tabla.
# En este primer caso, he querido probar con un E intermedio, para ello, he seleccionado la tabla 5. 
plot(Lista_de_Tablas_CHRM1[[5]]$length, Lista_de_Tablas_CHRM1[[5]]$evalue,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

```r
# Para poder estudiar de forma más clara y visual las diferencias entre un e más o menos restrictivo, he decido realizar la misma gráfica que la anterior, 
# analizando y comparando las mismas variables, pero, en este caso, con un e = 1, es decir, con la tabla 10, que corresponde al e más laxo.
plot(Lista_de_Tablas_CHRM1[[10]]$length, Lista_de_Tablas_CHRM1[[10]]$evalue,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

```r
# En este último caso, he querido comparar, la tabla 1 que corresponde al valor e más bajo. 
plot(Lista_de_Tablas_CHRM1[[1]]$length, Lista_de_Tablas_CHRM1[[1]]$evalue,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

```r
# Después de haber analizado los tres casos, podemos observar grandes diferencias entre ellos. Empezando por el caso más visual, vemos el resultado
# gráfico de la tabla 1, vemos muy pocos puntos y que corresponden a e del orden de -50. Estamos en un valor e muy restrictivo donde se presentan pocos
#alineamientos. En el caso del gráfico resultante de la tabla 10, tenemos para valores de e entorno a 1, muchos puntos, correspondientes a muchos 
# alineamientos.
```

```r
# Seguidamente, vamos a analizar las diferentes tablas, en este caso la 1, 5, 10. Esta gráfica muestra que la homología encontrada con valores e mayores 
# (tabla 10)no es aleatoria a lo largo de la secuencia de CHRM1, sino que, se concentra entre las posiciones ""400"", aproximadamente. En cambio, 
# En la tabla 1, con valores de e menores, vemos que se concentra a lo largo de todas las posiciones, esta diferencia tiene sentido. 
# Como hemos dicho anteriormente, si estamos en la tabla 1, estamos con un valor e muy restrictivo y con secuencias que son muy parecidas a la de mi proteína
# de interés, por ello, vemos que hay homologías a lo largo de todas las posiciones, en cambio en la 10, hay homologías en una posisición, ya que son valores
# de e mayores, más laxo. 
inicios <- Lista_de_Tablas_CHRM1[[1]]$qstart
finales <- Lista_de_Tablas_CHRM1[[1]]$qend
valoresE <- Lista_de_Tablas_CHRM1[[1]]$evalue

# Represento cada tabla: 1,5, 10; donde me analice la variable posiciones y la variable del valor e. La función segments() añade segmentos a un 
# gráfico previo.
AlineamientoMaximo_CHRM1 <- max(Lista_de_Tablas_CHRM1[[1]]$qend)
plot(c(0, AlineamientoMaximo_CHRM1), range(valoresE), type='n', xlab='Posición', ylab='Valor E')
segments(inicios, valoresE, finales, valoresE)
```

```r
inicios <- Lista_de_Tablas_CHRM1[[5]]$qstart
finales <- Lista_de_Tablas_CHRM1[[5]]$qend
valoresE <- Lista_de_Tablas_CHRM1[[5]]$evalue
AlineamientoMaximo_CHRM1 <- max(Lista_de_Tablas_CHRM1[[5]]$qend)
plot(c(0, AlineamientoMaximo_CHRM1), range(valoresE), type='n', xlab='Posición', ylab='Valor E')
segments(inicios, valoresE, finales, valoresE)
```

```r
# Finalmente, puedo con cada tabla representante de un valor de e determinado, pedirle al programa que me forme unas tablas,
# donde me indique para cada tabla, los grupos taxonómicos que se han encontrado. Aplico la función table() a esa columna de cada tabla en Lista_de_Tablas.
# A menor número e, más restrictivo, menor número de especies que coinciden con el alineamiento. Se observan en la (1), que la mayor coincidencia es con
# primates y roedores, que poseen un genoma más cercano evolutivamente. En cambio en la tabla 10, con un valor de e de 1, vemos muchas más especies y más
# dispares.
lapply(Lista_de_Tablas_CHRM1, function(x) table(x$sblastname))
```
