```r
# Como nos indica el enunciado, ahora vamos a analizar CHRNA3 y CHRNA7, que son dos subunidades parálogas
# del receptor nictonínico de acetilcolina. Según el artículo, estas proteínas aparecieron en el último 
# ancestro común entre cordados y cnidarios. Por lo que voy a proceder analizando como hemos hecho con
# CHRM1, ambas subunidades e iremos comentando las diferencias.
# Empezamos con CHRNA3:
BlastpOut_CHRNA3 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA3.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

TablaOut_CHRNA3 <- read.table(textConnection(BlastpOut_CHRNA3),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))

dim(TablaOut_CHRNA3)
```

```r
# Ahora veremos las dimsensiones de CHRNA7:
BlastpOutCHRNA7 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA7.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

TablaOutCHRNA7 <- read.table(textConnection(BlastpOutCHRNA7),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))

dim(TablaOutCHRNA7)
```

```r
# La primera diferencia que vemos ambas dimensiones de tablas, pero vamos a analizarlas más
# profundamente:
TablaOut_CHRNA3
```

```r
#En esta primera tabla, vemos que con un e muy pequeño, muy restrictivo, en esta tabla vemos muchas 
# coincidencias en los alineamientos en primates, nematodos, roedores, pájaros...
# Ahora analizamos CHRNA7:
TablaOutCHRNA7
```

```r
# En este caso, vemos también similitudes en roedores, nemátodos, primates... 
# Vamos a seguir el análisis para anotar más diferencias. Vamos a analizar tablas seriadas,
# iremos variando el valor de e, de más a menos restrictivo (e más pequeño a más grande) y estudiaremos
# las diferencias. A menor valor de e, más restrictivo, solo aquellos alineamientos que sean 
# muy similares se incluirán en la tabla, en cambio con el valor de e = 1, es mucho más laxo y veremos
# alineamientos más laxos. Por tanto, deberíamos observar más similitudes en la tabla 10 (e=1);
# que en la 1 (e=1xe^-50) donde es mucho más restrictivo.
# Empezamos con CHRNA3:
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

Lista_de_Tablas_CHRNA3 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut_CHRNA3 <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRNA3.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut_CHRNA3),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
#Cargamos también los valores de CHRNA7:
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')

Lista_de_Tablas_CHRNA7 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut_CHRNA7 <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRNA7.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut_CHRNA7),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
# En esta última tabla, observamos las mayores diferencias, mientras que en CHRNA7 no observamos
# una mayor diferencia, en CHRNA3 observamos que aparecen alineamientos con moscas, ascomicetos y 
# enterobacterias. 
# El siguiente paso es hacer una lista con todas las tablas, es decir; nosotros en las tablas de arriba
# solo veíamos los últimos alineamientos, por eso poníamos "tail", ahora vamos a poner en una lista,
# el nº de alineamientos en cada tabla, de izquierda a derecha tendremos: e menores (1) a e mayores (10):
Numero_de_resultados_CHRNA3 <- sapply(Lista_de_Tablas_CHRNA3, function(x) dim(x)[1])
Numero_de_resultados_CHRNA3
```
