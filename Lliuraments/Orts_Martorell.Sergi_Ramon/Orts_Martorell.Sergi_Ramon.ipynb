## Introducción
En este estudio pretendemos indagar en la taxonomia de las secuencias homologas a tres proteinas diferentes de interes, por un lado tendremos el receptor muscarinico de acetilcolina humano CHRM1, y las subunidades del receptor nicotinico de acetilcolina CHRNA3 y CHRNA7.
segun los estudios recientemente realizados se ha propuestos que estas proteinas aparecieron en etapas muy tempranas de a evolucion y que por tanto se deberian de encontrar en el ultimo ancestro comun entre los cordados y los cnidarios, excluyendo por completo plantas, hongos u otros organismos animales como las esponjas o los ctneoforos.

## Métodos
Con tal de corroborar estas suposiciones hemos llevado a cabo el siguiente estudio empleando las secuencias de las tres proteinas de interes juntamente con los paquetes swissprot y la herramienta de alineamiento blastp.

para ello empezamos generando el ambiente mediante el comando situado debao y generando tablas para la mejor visualizacion de las coincidencias, inicialmente emplearemos un valor E de 1.0e-50, este es un valor ciertamente restringido pero nuestro objetivo sera aumentar este valor con tal de observar como se expande la distribucion taxonomica.

```r
system2(command = './preparar_ambiente.sh', wait = TRUE)
```

```r

BlastpOut01 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRM1.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)


TablaOut01 <- read.table(textConnection(BlastpOut01),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))
dim(TablaOut01)
```

```r
TablaOut01
```

```r
# Observa entre los argumentes la especificación del formato de salida,
# "-outfmt". Si hace falta, desliza la barra para visualizar la línea completa.
BlastpOut02 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA3.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

# El resultado del comando anterior, guardado en "BlastOut01", es un montón de
# texto plano. Con las funciones textConnection() y read.table(), lo transformamos
# en un "data frame", una tabla donde cada columna es una variable. La opción
# "col.names" en read.table() sirve para dar nombre a las columnas.
TablaOut02 <- read.table(textConnection(BlastpOut02),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))
# Dimensiones de la tabla:
dim(TablaOut02)
```

```r
TablaOut02
```

```r
# Observa entre los argumentes la especificación del formato de salida,
# "-outfmt". Si hace falta, desliza la barra para visualizar la línea completa.
BlastpOut03 <- system2(command = 'blastp',
                      args = c('-db', 'swissprot',
                              '-query', 'CHRNA7.fas',
                              '-evalue', '1.0e-50',
                              '-outfmt', '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                      stdout = TRUE)

# El resultado del comando anterior, guardado en "BlastOut01", es un montón de
# texto plano. Con las funciones textConnection() y read.table(), lo transformamos
# en un "data frame", una tabla donde cada columna es una variable. La opción
# "col.names" en read.table() sirve para dar nombre a las columnas.
TablaOut03 <- read.table(textConnection(BlastpOut03),
                        sep = '\t',
                        col.names = c('saccver', 'pident', 'length', 'qstart',
                                     'qend', 'sstart', 'send', 'evalue', 'staxid',
                                     'ssciname', 'sblastname'))
# Dimensiones de la tabla:
dim(TablaOut03)
```

```r
TablaOut03
```

como hemos podido observar a pesar del restrictivo valor de E se han conseguido encontrar similitudes no solo entre vertebrados si no tambien en diversos invertebrados como insectos o nematodos.
Una vez observados por separado realizaremos los mismos estudios a diferentes valores de E cada vez mas permisivos. 

```r
Valores_E_maximos <- c('1.0e-50', '1.0e-40', '1.0e-30', '1.0e-20', '1.0e-10',
                       '1.0e-08', '1.0e-06', '1.0e-04', '1.0e-02', '1')
```

```r
Valores_E_maximos
```

```r
Lista_de_Tablas1 <- lapply(Valores_E_maximos,
                          function(x) {
                              BlastpOut <- system2(
                                  command = 'blastp',
                                  args = c('-db', 'swissprot',
                                           '-query', 'CHRM1.fas',
                                           '-evalue', x,
                                           '-outfmt',
                                           '"7 saccver pident length qstart qend sstart send evalue staxid ssciname sblastname"'),
                                  stdout = TRUE)
                              read.table(textConnection(BlastpOut),
                                         sep = '\t',
                                         col.names = c('saccver', 'pident', 'length', 'qstart',
                                            'qend', 'sstart', 'send', 'evalue', 'staxid',
                                            'ssciname', 'sblastname'))
                          })
```

```r
tail(Lista_de_Tablas1[[1]])
```

```r
Numero_de_resultados1 <- sapply(Lista_de_Tablas1, function(x) dim(x)[1])
Numero_de_resultados1
```

```r
plot(Valores_E_maximos, Numero_de_resultados1, log = 'xy', type = 'b')
```

Como era de esperar a medida que aumentamos el valor de E empezamos a ver un crecimiento exponencial de las coincidencias encontradas.
Acto seguido estudiaremos la relacion entre la longitud de los fragmentos y el valor E determinado, para ello emplearemos la ultima tabla generada siendo esta la que posee el valor E mas laxo y por tanto la que mayor numero de coincidencias presenta.

```r
plot(Lista_de_Tablas1[[10]]$length, Lista_de_Tablas1[[10]]$evalue,
     xlab = 'Longitud de alineamiento', ylab = 'Valor E')
```

```r
inicios1 <- Lista_de_Tablas1[[10]]$qstart
finales1 <- Lista_de_Tablas1[[10]]$qend
valoresE1 <- Lista_de_Tablas1[[10]]$evalue

# La función segments() añade segmentos a un gráfico previo. Por eso, represento
# antes un gráfico con los rangos adecuados, pero vacío (type='n'). Para determinar
# el rango del eje horizontal, determino primero el valor máximo de "qend".
AlineamientoMaximo1 <- max(Lista_de_Tablas1[[10]]$qend)
plot(c(0, AlineamientoMaximo1), range(valoresE1), type='n', xlab='Posición', ylab='Valor E')
segments(inicios1, valoresE1, finales1, valoresE1)
```
